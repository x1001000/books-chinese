```
import requests
from bs4 import BeautifulSoup
```

```
url = 'https://www.books.com.tw/web/books/?loc=menu_1_001'
r = requests.get(url)
print(r.status_code)
print(r.text)
```

```
soup = BeautifulSoup(r.text)
soup
```

```
new_books = soup.find_all(class_='li_M201101_17350__P00a400010055 item')
new_books
```

```
for book in new_books:
    print(f'書名：{book.text.split()[1]}')
    print(book.text.split()[2].split('優惠價')[0])
    #for item in book.text.split()[1:3]:
    #    print(item)
    print(book.find('a').get('href'))
    #break
```
